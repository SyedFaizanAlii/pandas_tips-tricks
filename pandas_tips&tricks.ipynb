# 03- How to Rename Data Columns

```python
import pandas as pd
df = pd.DataFrame ({ 'A Col' : [1,2,3,4,5,6], 'B Col' : [7,8,9,10,12,11]})
#print(df)

df
```

```python
# 1st trick to rename columns name
df.rename(columns={'A' : 'Name', 'B':'Marks'}, inplace=True)
df
```

```python
# 2nd trick to rename columns name
df.columns=['Col_aa','Col_bb']
df
```

```python
# to replace any string and character
df.columns=df.columns.str.replace(' ','*')
df
```

```python
#Adding prefix to columns
df = df.add_prefix('baba_')
df
```

```python
#Adding suffix to columns
df = df.add_suffix('_baba')
df
```
